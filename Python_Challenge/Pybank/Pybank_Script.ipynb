```python
#import os and csv
import os
import csv

# Path to collect data from the Resources folder
datapath = os.path.join('Resources', 'BudgetData.csv')
```

```python
# Define the function and have it accept the 'budget_data' as its sole parameter
def print_percentages(budget_data):

    # For readability, it can help to assign your values to variables with descriptive names
    total_months = 0
    total_profit = 0
    average_change = 0
    greatest_increase = 0
    greatest_decrease = 0
    previous_month = 0
    current_month = 0
    change = 0
    total_change = 0
    average_change = 0
    month_of_change = []

    # Read in the CSV file
    with open(budget_data, 'r') as csvfile:

        # Split the data on commas
        csvreader = csv.reader(csvfile, delimiter=',')

        # Skip the header labels to iterate with the values
        header = next(csvreader)

        # Loop through the data
        for row in csvreader:

            # Track the total
            total_months = total_months + 1

            # Track the profit
            total_profit = total_profit + int(row[1])

            # Track the changes
            current_month = int(row[1])
            change = current_month - previous_month
            month_of_change = month_of_change + [row[0]]
            total_change = total_change + change
            previous_month = current_month

            # Calculate the greatest increase
            if change > greatest_increase:
                greatest_increase = change
                greatest_increase_month = row[0]

            # Calculate the greatest decrease
            if change < greatest_decrease:
                greatest_decrease = change
                greatest_decrease_month = row[0]

        # Calculate the Average & The Date
        average_change = total_change / (total_months - 1)

        # Generate Output Summary
        print(f"Financial Analysis")
        print(f"----------------------------")
        print(f"Total Months: {total_months}")
        print(f"Total: ${total_profit}")
        print(f"Average Change: ${average_change:.2f}")
        print(f"Greatest Increase in Profits: {greatest_increase_month} (${greatest_increase})")
        print(f"Greatest Decrease in Profits: {greatest_decrease_month} (${greatest_decrease})")

        #export a text file with the results   
        output_file = os.path.join("Analysis", "PyBank.txt")
        with open(output_file, "w") as txtfile:
            txtfile.write(f"Financial Analysis")
            txtfile.write(f"----------------------------")
            txtfile.write(f"Total Months: {total_months}")
            txtfile.write(f"Total: ${total_profit}")
            txtfile.write(f"Average Change: ${average_change:.2f}")
            txtfile.write(f"Greatest Increase in Profits: {greatest_increase_month} (${greatest_increase})")
            txtfile.write(f"Greatest Decrease in Profits: {greatest_decrease_month} (${greatest_decrease})")
```
